```python
import pandas as pd
```

```python


dfCR = pd.read_csv('dataCSV/TASS2019_country_CR_train.csv')
dfMX = pd.read_csv('dataCSV/TASS2019_country_MX_train.csv')
dfPE = pd.read_csv('dataCSV/TASS2019_country_PE_train.csv')
dfUY = pd.read_csv('dataCSV/TASS2019_country_UY_train.csv')

```

```python
df = pd.concat([dfCR, dfMX, dfPE, dfUY], ignore_index=True)
```

```python
df = df[df['sentiment'].notnull()]
df
```

```python


from nltk.corpus import stopwords
import re

```

```python
stop_words = set(stopwords.words('spanish'))
len(stop_words)
```

```python


def clean_text(text):
    text = re.sub(r'@[A-Za-z09]+', '', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'_', ' ', text)
    text = re.sub(r'\d', '', text)
    return text

df['content'] = df['content'].apply(clean_text)
df

```

```python


from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer(stop_words=stop_words)
print(count_vector)

```

```python


documents = df['content'].to_list()
documents

```

```python


count_vector.fit(documents)
names = count_vector.get_feature_names()
names

```

```python


doc_array = count_vector.transform(documents).toarray()
doc_array



```

```python



frequency_matrix = pd.DataFrame(data=doc_array, columns=names)
frequency_matrix



```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['content'], df['sentiment'], random_state=1)
print('Number of rows in the total set: {}'.format(df.shape[0]))
print('Number of rows in the training set: {}'.format(X_train.shape[0]))
print('Number of rows in the test set: {}'.format(X_test.shape[0]))
```

```python


# Instantiate the CountVectorizer method
count_vector = CountVectorizer(stop_words=stop_words)
# Fit the training data and then return the matrix
training_data = count_vector.fit_transform(X_train)
# Transform testing data and return the matrix. Note we are not fitting the testing data into the CountVectorizer()
testing_data = count_vector.transform(X_test)

```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=13)
knn_model.fit(training_data, y_train)
```

```python
predictions = knn_model.predict(testing_data)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Accuracy score: ', format(accuracy_score(y_test, predictions)))
print('Precision score: ', format(precision_score(y_test, predictions, average=None)))
print('Recall score: ', format(recall_score(y_test, predictions, average=None)))
print('F1 score: ', format(f1_score(y_test, predictions, average=None)))
```

```python



import pickle
pickle.dump(count_vector, open('models/knn/count_vectorKNN.sav', 'wb'))
pickle.dump(knn_model, open('models/knn/classifier_KNN.sav', 'wb'))

```
